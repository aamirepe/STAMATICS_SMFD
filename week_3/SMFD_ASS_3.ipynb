```python
import numpy as np
import matplotlib.pyplot as plt
```

* QUESTION 1
  

```python
# Parameters
T = 1.0       # total time
N = 1000      # number of steps
dt = T / N    # time step
t = np.linspace(0, T, N+1)

# Standard Brownian motion: W(0) = 0
W = np.zeros(N+1)
W[1:] = np.cumsum(np.sqrt(dt) * np.random.randn(N))

# Plot
plt.plot(t, W)
plt.title('Standard Brownian Motion Path')
plt.xlabel('Time t')
plt.ylabel('W(t)')
plt.grid()
plt.show()
```

 * QUESTION 2
   

```python
# Parameters
alpha = 0.1
sigma = 0.2
S0 = 100
num_paths = 5

plt.figure(figsize=(10, 6))

for _ in range(num_paths):
    W = np.zeros(N+1)
    W[1:] = np.cumsum(np.sqrt(dt) * np.random.randn(N))
    S = S0 * np.exp(sigma * W + (alpha - 0.5 * sigma**2) * t)
    plt.plot(t, S)

plt.title('5 Paths of Geometric Brownian Motion')
plt.xlabel('Time t')
plt.ylabel('S(t)')
plt.grid()
plt.show()
```

### Q3. Show that for standard Brownian motion, \( \mathbb{E}[W_s W_t] = \min(s, t) \) for \( s, t \geq 0 \)

Assume without loss of generality that \( s \leq t \).  
We can write the Brownian motion at time \( t \) as:
\[
W_t = W_s + (W_t - W_s)
\]

Now compute the expectation:
\[
\mathbb{E}[W_s W_t] = \mathbb{E}[W_s (W_s + (W_t - W_s))] = \mathbb{E}[W_s^2] + \mathbb{E}[W_s (W_t - W_s)]
\]

Since \( W_t - W_s \) is independent of \( W_s \), and has mean 0:
\[
\mathbb{E}[W_s (W_t - W_s)] = \mathbb{E}[W_s] \cdot \mathbb{E}[W_t - W_s] = 0
\]

Also, \( \mathbb{E}[W_s^2] = \text{Var}(W_s) = s \)

Therefore:
\[
\mathbb{E}[W_s W_t] = s = \min(s, t)
\]
